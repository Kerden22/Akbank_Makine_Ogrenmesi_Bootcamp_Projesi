```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Veri setini inceleme

```python
# Veri setini yükle
data = r"C:\Users\acer\Desktop\Credit_Card_Fraud_Detection_ML\creditcard.csv"
df = pd.read_csv(data)

# İlk 20 satırı göster
df.head(5)
```

```python
# Veri setinin genel bilgileri
df.info()
print("----------------------------------------------")
print("----------------------------------------------")
print("----------------------------------------------")
# Eksik değerlerin sayısını kontrol et
df.isnull().sum()
```

```python
# Fraud işlemler çok az, veri dengesiz mi?
# Veri seti dengesiz (imbalanced) olduğu için normal işlemler (0) çok fazla, 
# dolandırıcılık işlemleri (1) çok az görünecek.  (Genellikle %0.172 fraud oranı var.)
# Class değişkenine göre işlem sayısını görselleştirelim
plt.figure(figsize=(6,4))
sns.countplot(x=df["Class"], hue=df["Class"], palette=["green", "red"], legend=False)
plt.title("Normal İşlemler (0) vs Dolandırıcılık (1)")
plt.xlabel("İşlem Türü")
plt.ylabel("İşlem Sayısı")
plt.show()

# Sınıf dağılımını yüzdesel olarak görelim
fraud_ratio = df["Class"].value_counts(normalize=True) * 100
print(f"Sınıf Dağılımı:\n{fraud_ratio}")
```

```python
# Fraud işlemler belli saatlerde mi yoğunlaşıyor?
# Eğer dolandırıcılıklar belli saat aralıklarında yoğunlaşmışsa, bu bizim için önemli bir analiz olabilir.
# Örneğin, gece saatlerinde dolandırıcılık işlemleri daha fazla mı oluyor?
plt.figure(figsize=(10,5))
sns.histplot(df[df["Class"]==0]["Time"], bins=50, color="green", label="Normal İşlemler", kde=True)
sns.histplot(df[df["Class"]==1]["Time"], bins=50, color="red", label="Dolandırıcılık", kde=True)
plt.title("Zaman Değişkeninin Dolandırıcılık Üzerindeki Etkisi")
plt.xlabel("Zaman (saniye)")
plt.ylabel("İşlem Sayısı")
plt.legend()
plt.show()
```

```python
# Bu grafik sayesinde yalnızca dolandırıcılık işlemlerinin zaman dağılımını net bir şekilde görebiliriz.
plt.figure(figsize=(10,5))
sns.histplot(df[df["Class"]==1]["Time"], bins=50, color="red", kde=True)
plt.title("Dolandırıcılık İşlemlerinin Zaman Dağılımı")
plt.xlabel("Zaman (saniye)")
plt.ylabel("Dolandırıcılık İşlem Sayısı")
plt.show()
```

```python
# Hangi özellikler birbirine bağlı?
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), cmap="coolwarm", annot=False, linewidths=0.5)
plt.title("Korelasyon Matrisi")
plt.show()
```
